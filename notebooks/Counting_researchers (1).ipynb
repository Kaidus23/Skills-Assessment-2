**(2) Importing Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
```

**(3) Count publications per author and visualize the top contributors**

```python
# Load the authors data
authors = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Assesment/Skills for Data Scientist/Group 4/authors.APOE.csv")

# Count publications by the author
author_counts = authors.groupby(['AuthorLastname', 'AuthorInitials']).size().reset_index(name='Count')
top_authors = author_counts.nlargest(10, 'Count')

# Visualization
plt.figure(figsize=(10, 6))
#disabling the error bar by adding the errorbar=None argument in the sns.barplot function
sns.barplot(y='AuthorLastname', x='Count', data=top_authors, hue='AuthorLastname', dodge=False, legend=False, errorbar=None)
plt.title('Top 10 Most Active Researchers')
plt.xlabel('Number of Publications')
plt.ylabel('Author')
plt.legend([], [], frameon=False)
plt.show()

```
