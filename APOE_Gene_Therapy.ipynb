**(2) Importing Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
```

**(3) Loading the data and visualizing publication trends**

```python
# Load the datas
articles = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Assesment/Skills for Data Scientist/Group 4/articles.APOE.csv")
paper_counts = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Assesment/Skills for Data Scientist/Group 4/paper_counts.csv")

# Publication trends
articles_per_year = articles.groupby('Year').size().reset_index(name='Count')

# Merge with total papers
trend_data = pd.merge(articles_per_year, paper_counts, on='Year', how='left')
trend_data['Proportion'] = trend_data['Count_x'] / trend_data['Count_y']

# Visualization
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Count_x', data=trend_data, label='APOE-related Papers')
sns.lineplot(x='Year', y='Proportion', data=trend_data, label='Proportion of APOE Papers')
plt.title('Publication Trends')
plt.xlabel('Year')
plt.ylabel('Count / Proportion')
plt.legend()
plt.show()
```
