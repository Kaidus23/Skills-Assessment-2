**(3)Loading the data and visualizing Geographical Distribution of Research**

```python
# Load the dataset (already loaded, using 'data' from the uploaded file)
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assesment/Skills for Data Scientist/Group 4/articles.APOE.csv')

# Count the number of papers per journal
journal_counts = df['Journal'].value_counts()

# Plot the top 10 journals with the highest number of publications
journal_counts.head(10).plot(kind='bar', figsize=(10, 6), color='red', edgecolor='black')
plt.title('Top 10 Journals by Number of Publications', fontsize=16)
plt.xlabel('Journal', fontsize=14)
plt.ylabel('Number of Publications', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(alpha=0.5)
plt.show()
```
